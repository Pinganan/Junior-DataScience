```python
# %load UI.py

from calculateWeight import Cal_Model

import tkinter as tk

class Windows(object):

    def __init__(self):
        self.window = tk.Tk()
        self.img = None
        self.set_music_version()
        self.selection()
        self.radiobutton1.select()

    # music chatbot UI
    def set_music_version(self):
        self.window.title('歌曲查詢機器人')
        self.window.configure(background='LightYellow')

        # lyrcs, feeling radiobutton
        self.var1 = tk.IntVar()
        self.radiobutton1 = tk.Radiobutton(self.window, text = "歌詞", font=16, background='LightYellow', variable = self.var1, value = 3, command = self.selection)
        self.radiobutton1.grid(column=0, row=0, padx=10, sticky=tk.W)
        self.radiobutton2 = tk.Radiobutton(self.window, text = "類別", font=16, background='LightYellow', variable = self.var1, value = 0, command = self.selection)
        self.radiobutton2.grid(column=0, row=1, padx=10, sticky=tk.W)
        # choice label
        self.choice_label = tk.Label(self.window, text = "", font=16, background='LightYellow')
        self.choice_label.grid(column=0, row=2, padx=10, sticky=tk.W)
        # input text field
        self.text = tk.Text(self.window, font=16, bd=2, width=75, height=10)
        self.text.grid(column=0, row=3, columnspan=2, padx=15, sticky=tk.N+tk.W)
        # sumit button
        self.button1 = tk.Button(self.window, text = "確定", command = self.sumbit_user_input)
        self.button1.grid(column=2, row=3, padx=10, pady=5)
        # result text field
        self.ans_label1 = tk.Label(self.window, bd=2, relief="ridge", font=16, background='FloralWhite', width=42, height=23, justify='left', wraplength=500, text="歌名")
        self.ans_label1.grid(column=0, row=4, padx=15, pady=15, sticky=tk.W+tk.N)
        # starting img
        self.img = tk.PhotoImage(file = 'cat_search.gif')
        self.img_label1 = tk.Label(self.window, image = self.img, background='LightYellow')
        self.img_label1.grid(column=1, row=4, columnspan=2, sticky=tk.S+tk.E)

    # select type or feeling for question
    def selection(self):
        if self.var1.get() == 0:
            selstr = "類別"
        else:
            selstr = "歌詞"

        self.choice_label.config(text='請輸入' + str(selstr) + '：')

    # model calculating result on UI
    def sumbit_user_input(self):
        r = Cal_Model(question_num=self.var1.get()).compare_data(self.text.get("1.0", "end"))

        # reset UI screen
        self.ans_label1 = tk.Label(self.window, bd=2, relief="ridge", font=16, background='FloralWhite', width=42, height=23, justify='left', wraplength=500, text="")
        self.ans_label1.grid(column=0, row=4, padx=15, pady=15, sticky=tk.W+tk.N)

        if not r:
            self.img = tk.PhotoImage(file = 'cat_fail.gif')
            self.img_label1 = tk.Label(self.window, image = self.img, background='LightYellow')
            self.img_label1.grid(column=1, row=4, columnspan=2, sticky=tk.S+tk.E)
            self.ans_label1['text'] = "未查詢到相關結果"
            return

        s = ""
        count = 1
        for i in r:
            if count == 9:
                s = s + str(count + 1) + ". " + str(i)
                break
            s = s + str(count) + ". " + str(i) + "\n\n"
            count = count + 1

        self.img = tk.PhotoImage(file = 'cat_result.gif')
        self.img_label1 = tk.Label(self.window, image = self.img, background='LightYellow')
        self.img_label1.grid(column=1, row=4, columnspan=2, sticky=tk.S+tk.E)
        self.var2 = tk.StringVar()
        self.ans_label1['text'] = ""
        self.ans_label1 = tk.Label(self.window, bd=0, relief="ridge", font=16, background='FloralWhite', textvariable=self.var2, justify='left', wraplength=500)
        self.ans_label1.grid(column=0, row=4, padx=50, pady=60, sticky=tk.W+tk.N)
        self.var2.set(s)

if __name__ == '__main__':
    Windows().window.mainloop()
```
